## Home Price Prediction Model

`In this Simple Linear Regression`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("../DATASETS/HomePrices.csv")
df
```

```python
%matplotlib inline
plt.scatter(df["Area"], df["Price"], color="red", marker="+")
plt.xlabel("Area (in sq ft)")
plt.ylabel("Price (in USD)")
plt.title("Data Plot")
plt.grid()
plt.show()
```

ðŸ‘‡Here implimenting Linear Regression â€”â€”â€” (y = mx + c)

```python
reg = linear_model.LinearRegression()
reg.fit(df[["Area"]], df.Price)
```

```python
reg.predict([[2000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
213.77344612*2000+(-5086.428834130173)
```

```python
d = pd.read_csv("../DATASETS/Areas.csv")
```

```python
p = reg.predict(d)
```

```python
%matplotlib inline
plt.xlabel("Area (in sq ft)")
plt.ylabel("Price (in USD)")
plt.title("Data & Prediction Plot")
plt.scatter(df.Area, df.Price, color="red", marker="+")
plt.plot(df.Area, reg.predict(df[["Area"]]), color="blue") 
plt.grid()
plt.show()
```

```python
d["prices"] = p
```

```python
d.to_csv("../DATASETS/Areas.csv", index=False)
```
