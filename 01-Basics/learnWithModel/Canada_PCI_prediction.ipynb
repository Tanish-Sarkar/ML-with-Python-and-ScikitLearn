```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("../DATASETS/canada_per_capita_income.csv")
df.head(20)
```

```python
%matplotlib inline
plt.scatter(df["year"], df["pci"], color="red", marker="+")
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Data Plot")
plt.grid()
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["year"]], df.pci)
```

```python
reg.predict([[2000]])
```

```python
d = pd.read_csv("../DATASETS/pciTest.csv")
```

```python
p = reg.predict(d)
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Data & Prediction Plot")
plt.scatter(df.year, df.pci, color="red", marker="+")
plt.plot(df.year, reg.predict(df[["year"]]), color="blue") 
plt.grid()
plt.show()
```

```python
d["pci"] = p
d.to_csv("../DATASETS/pciPrediction.csv", index=False)
```
