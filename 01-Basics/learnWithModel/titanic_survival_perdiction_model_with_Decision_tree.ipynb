### **`Titanic Survival Prediction Model With Decision Tree`**

In this file using following columns build a model to predict if person would survive or not,
- Pclass
- Sex
- Age
- Fare

**Calculate score of your model**

Loading the Dataset

```python
import pandas as pd
df = pd.read_csv("../../DATASETS/titanic.csv")
df.head()
```

cleaning the dataset and by taking the required features and leaving the rest

```python
df_required = df[['Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
df_required.head(3)
```

Converting the feature 'Sex' into dummy variable "sex_n"

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
```

```python
df_required["sex_n"] = le_sex.fit_transform(df_required["Sex"])
df_required
```

EDA

```python
df_required["Age"] = df_required["Age"].fillna(df_required["Age"].median())
df_required
```

```python
df_required.isna().count()
```

```python
df_required = df_required.drop(["Sex"], axis="columns")
df_required
```

taking the input and targets in the form of X & y


```python
X = df_required.drop(["Survived"], axis="columns")
y = df_required["Survived"]
```

train and test model split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=43)
```

`trainning the decision tree model`

```python
from sklearn.tree import DecisionTreeClassifier

dtree_model = DecisionTreeClassifier()
dtree_model.fit(X_train, y_train)
y_pred = dtree_model.predict(X_test)
```

```python
dtree_model.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import (r2_score, mean_squared_error)

dtree_model_reg = DecisionTreeRegressor()
dtree_model_reg.fit(X_train, y_train)
y_pred = dtree_model_reg.predict(X_test)

score = r2_score(y_test, y_pred)
print(score)
score2 = mean_squared_error(y_test, y_pred)
print(score2)
```
