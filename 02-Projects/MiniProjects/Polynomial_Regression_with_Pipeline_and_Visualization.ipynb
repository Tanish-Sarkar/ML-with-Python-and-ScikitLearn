### Polynomial Regression with Pipeline + Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
# Generate some quadratic-like data
np.random.seed(1)
X = np.random.rand(100, 1) * 10
y = 3*X**2 + 2*X + 5 + np.random.randn(100, 1) * 10   # y = 3x² + 2x + 5 + noise
```

```python
# Degree 2 Polynomial Regression Pipeline
model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model.fit(X, y)
y_pred = model.predict(X)
```

```python
# Evaluate performance
print(f"R² score: {r2_score(y, y_pred)}")
```

```python
# Visualization
plt.scatter(X, y, color='black', label='Actual Data')
plt.scatter(X, y_pred, color='red', label='Predictions')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression (Degree 2)")
plt.legend()
plt.grid(True)
plt.show()
```
