```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

MSE

```python
%matplotlib inline

def gradient_descent(x, y):
    m_curr = b_curr = 0
    learning_rate = 0.01
    n = len(x)
    
    plt.scatter(x, y, color='red', marker='+', linewidth=5)
    
    for i in range(10000):
        y_predicted = m_curr * x + b_curr
        plt.plot(x, y_predicted, color='green', alpha=0.1)  # Added alpha for transparency
        
        # Calculate gradients
        md = -(2/n) * np.sum(x * (y - y_predicted))  # Derivative w.r.t. m
        bd = -(2/n) * np.sum(y - y_predicted)        # Derivative w.r.t. b
        
        # Update parameters
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
    
    # Print final values
    print(f"Final slope (m): {m_curr}, Final intercept (b): {b_curr}")
```

```python
data = pd.read_csv("../../DATASETS/test_scores.csv")
x = data["math"].to_numpy()  # Corrected typo: 'tonumpy' → 'to_numpy'
y = data["cs"].to_numpy()

# Run gradient descent
gradient_descent(x, y)
plt.show()  # Display the plot
```

MAE(Mean Absolute Error) and R^2

```python
from sklearn.metrics import r2_score  # For R-squared calculation

%matplotlib inline

def gradient_descent(x, y):
    m_curr = b_curr = 0
    learning_rate = 0.01
    n = len(x)
    
    plt.scatter(x, y, color='red', marker='+', linewidth=5)
    
    for i in range(10000):
        y_predicted = m_curr * x + b_curr
        plt.plot(x, y_predicted, color='green', alpha=0.1)
        
        # Gradients
        md = -(2/n) * np.sum(x * (y - y_predicted))
        bd = -(2/n) * np.sum(y - y_predicted)
        
        # Update parameters
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
    
    plt.title("Gradient Descent")
    # Final predictions
    y_final = m_curr * x + b_curr
    
    # Calculate MAE
    mae = np.mean(np.abs(y - y_final))
    
    # Calculate R-squared
    # r2 = r2_score(y, y_final)
    ss_res = np.sum((y - y_final) ** 2)      # Sum of squared residuals
    ss_tot = np.sum((y - np.mean(y)) ** 2)   # Total sum of squares
    r2 = 1 - (ss_res / ss_tot)
    
    print(f"Final slope (m): {m_curr:.4f}, Final intercept (b): {b_curr:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"R-squared (R²): {r2:.4f}")

# Load data
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

# Run gradient descent
gradient_descent(x, y)
plt.show()
```
